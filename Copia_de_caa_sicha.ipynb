````
## caso ozono
VOL_OZONO =  df_rename_3['ozono_A']*5/4095
RS_OZONO =  1000*((5-VOL_OZONO)/VOL_OZONO )
OZONO = 33.258*pow(RS_OZONO/4895, -1.959)
OZONO = OZONO/1000
K = 0.082*(df_rename_3['temperatura']+273.15)/(df_rename_3['presion_atm']/1013.25)
df_rename_3['ozono_A'] = OZONO * 48*1000 / (K)
df_rename_3
````

````
## caso sharp 
raw_sharp = df_rename_3['SHARP_A']
vol_sharp = raw_sharp*5/(4095)
df_rename_3['SHARP_A'] = ((0.17*vol_sharp-0.1)*100).round(2)
print(vol_sharp.describe())
df_rename_3['SHARP_A'].describe()
````

````
 #CASO SO2
 #en este caso divido entre 10000 debido a que las lecturas son mas reales
 df_rename_3['SO2_A'] = ( (((df_rename_3['SO2_A']*499/4095) + 1 )*64*1000/K)/10000 ).round(2)
 df_rename_3['SO2_A']
````

````
# caso radiacion 
df_rename_3['radiacion'] = ( 4.7082*(df_rename_3['radiacion']*1024/4095) - 23.651 ).round(2)
df_rename_3['radiacion']
````

````
## arreglo en el NO2
  # float NO2_raw = analogRead(CMJU_NO2); //no2  33
  # float NO2_conv = 0.05+ NO2_raw*((5-0.05)/1024);
  # float K = 0.082*(t2+273)/(pre*100/101325) ;
  # float NO2 = NO2_conv *( 48*100 / K ) ;
k_no2 = 0.082*(df_rename_3['temperatura']+273.15)/(df_rename_3['presion_atm']/1013.25)
no2_raw = ((df_rename_3['NO2_A']*k_no2/4800)-0.05)*1024/4.95
# recalculando
no2_raw_esp32 = no2_raw
no2_conv = 0.05 + no2_raw_esp32*((10-0.05)/4095)

#observando la informacion es demasiado alto, aplicaro una division entre 10
#con la finalidad de hacer mas creible la informacion
df_rename_3['NO2_A'] = ( (no2_conv*(32+14)*1000/k_no2)/100 ).round(2)
df_rename_3['NO2_A'] 
````

````
#@title mapdisplay: Crea mapas interactivos usando folium
import folium
def mapdisplay(center, dicc, Tiles="OpensTreetMap",zoom_start=10):
    '''
    :param center: Center of the map (Latitude and Longitude).
    :param dicc: Earth Engine Geometries or Tiles dictionary
    :param Tiles: Mapbox Bright,Mapbox Control Room,Stamen Terrain,Stamen Toner,stamenwatercolor,cartodbpositron.
    :zoom_start: Initial zoom level for the map.
    :return: A folium.Map object.
    '''
    center = center[::-1]
    mapViz = folium.Map(location=center,tiles=Tiles, zoom_start=zoom_start)
    for k,v in dicc.items():
      if ee.image.Image in [type(x) for x in v.values()]:
        folium.TileLayer(
            tiles = v["tile_fetcher"].url_format,
            attr  = 'Google Earth Engine',
            overlay =True,
            name  = k
          ).add_to(mapViz)
      else:
        folium.GeoJson(
        data = v,
        name = k
          ).add_to(mapViz)
    mapViz.add_child(folium.LayerControl())
    return mapViz
````

````
#@title aplicando ecuaciones a las imagenes
# datos de la ecuacion
# Intercept    38.041346
# NO2_S5       -0.165507
#y=-0.166*x38.041
no2_clip = NO2_COLECION.select('NO2_column_number_density').mean().clip(lurigancho)
no2_clip_ecuacion = (no2_clip.multiply(1e6).multiply(-0.166)).add(38.041346)
no2_clip_ecuacion
no2_clip_inca = no2_clip_ecuacion.multiply(100).divide(50)

# datos de la ecuacion
# Intercept    56.002928
# SO2_S5        0.000615
# ecuacion
# y=0.001*x56.003

so2_clip = SO2_COLECCION.select('SO2_column_number_density').mean().clip(lurigancho)
so2_clip_ecuacion = (so2_clip.multiply(0.000616)).add(56.002928)
so2_clip_ecuacion
so2_clip_inca = so2_clip_ecuacion.multiply(100).divide(250)

# datos de la ecuacion
# Intercept   -1943.005289
# OZONO_S5        0.018623
# ecuacion
# y=0.019*x-1943.005

ozono_clip = OZONO_COLECCION.select('O3_column_number_density').mean().clip(lurigancho)
ozono_clip_ecuacion = (ozono_clip.multiply(0.018623)).subtract(1943.005289)
ozono_clip_ecuacion
ozono_clip_ndvi = ozono_clip_ecuacion.multiply(100).divide(100)

# datos de la ecuacion
# Intercept     3.308639
# AEROSOL_S5   -0.000005
# ecuacion
# y=-0.0*x3.309

pm2_5_clip = AEROSOL_COLECCION.select('absorbing_aerosol_index').mean().clip(lurigancho)
pm2_5_clip_ecuacion = (pm2_5_clip.multiply(-0.000005)).add(3.308639)
pm2_5_clip_ecuacion
pm2_5_clip_inca = pm2_5_clip_ecuacion.multiply(100).divide(50)
````
